```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=cv2.imread("zebra.jpg")
```

```python
plt.imshow(image)
plt.show()
```

This image is in BGR format

```python
image_rgb=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

```python
image_hsv=cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
plt.imshow(image_hsv)
plt.axis('off')
plt.show()
```

```python
lower_white=np.array([0,0,190]) 
upper_white=np.array([30,80,255]) # there is some level of brown
```

```python
# color targetting

mask=cv2.inRange(image_hsv, lower_white, upper_white)
```

```python
plt.imshow(mask)
plt.axis('off')
plt.show()
```

```python
black_hsv=np.uint8([[[190,255,0]]]) # hue=190, saturation=255, brightness=0, can take any other color
black_bgr=cv2.cvtColor(black_hsv, cv2.COLOR_HSV2BGR)[0][0] # want only particular color
```

```python
image_result=image.copy()
image_result[mask>0]=black_bgr # black out the white area in the original image
```

```python
image_result_rgb=cv2.cvtColor(image_result, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image_result_rgb)
plt.axis('off')
plt.show()
```

```python
# plotting the result
fig,ax=plt.subplots(1,3,figsize=(12,6))
ax[0].imshow(image_rgb)
ax[0].set_title('Original Image')
ax[0].axis('off')

ax[1].imshow(mask)
ax[1].set_title('Original Image')
ax[1].axis('off')

ax[2].imshow(image_result_rgb)
ax[2].set_title('Original Image')
ax[2].axis('off')

plt.tight_layout()
plt.show()
```

In the last image we can see that the white stripes has been replaced with black color. All the whites has been replaced by black color.
